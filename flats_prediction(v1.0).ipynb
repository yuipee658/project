Исследую данные:

```python
print(dataoutnan['region_of_moscow'].value_counts())
print(dataoutnan['is_new'].value_counts())
print(dataoutnan['is_apartments'].value_counts())
```

Замечу, что "region_of_moscow", "is_new", "is_apartments" являются категориальными признаками; "is_new", "is_apartments" могут принимать только значения "0" или "1", с ними все хорошо. Для переменной "region_of_moscow" применим One-Hot Encoding:

```python

print("Значения region_of_moscow:")
print(dataoutnan['region_of_moscow'].value_counts())
embarked_encoded=pnd.get_dummies(dataoutnan['region_of_moscow'], prefix='region_of_moscow')
print(f"\nOne-Hot Encoding создал столбцы: {list(embarked_encoded.columns)}")
print("\nПервые 5 строк:")
print(embarked_encoded.head())
dataset_1=pnd.concat([dataoutnan['region_of_moscow'], embarked_encoded], axis=1)
```

```python
print(dataset_1.value_counts())
```

Проверяю данные на выбросы, построив для них boxplots:

```python
import seaborn as sbn
for column in dataoutnan.columns[1:6]:
    mpl.figure(figsize=(6, 4))
    sbn.boxplot(data=dataoutnan, y=column)
    mpl.title(f"Boxplot for {column}")
    mpl.show()
```

Наконец-то создаю и обучаю модель:

```python
from sklearn.preprocessing import StandardScaler
y=dataoutnan['price']
X=dataoutnan.drop(columns = ['price'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print(X_train)
model = LinearRegression()
model.fit(X_train, y_train)
```

Замерю R2:

```python
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
print("R2:", r2)
```
